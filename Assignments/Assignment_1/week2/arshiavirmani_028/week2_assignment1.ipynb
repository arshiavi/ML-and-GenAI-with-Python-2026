```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Q1. Load the dataset and display the first five records
df = pd.read_csv('ShopSmart_Customers.csv')
print("\nQ1. First 5 records of ShopSmart dataset:")
print(df.head())

# Q2. Determine the number of rows and columns
print(f"\nQ2. Dataset Shape (Rows, Columns): {df.shape}")

# Q3. Display all column names
print("\nQ3. Column names:")
print(df.columns.tolist())

# Q4. Identify numerical and categorical features
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"\nQ4. Numerical features: {numerical_cols}")
print(f"    Categorical features: {categorical_cols}")

# Q5. Check whether the dataset contains missing values
print("\nQ5. Missing values per column:")
print(df.isnull().sum())

# Q6. Calculate the average age of users
print(f"Q6. Average age of customers: {df['Age'].mean():.2f} years")

# Q7. Determine the average time spent on site (Proxy for watch hours)
print(f"Q7. Average time spent on site: {df['TimeOnSite'].mean():.2f} minutes")

# Q8. Find the average spending of users
print(f"Q8. Average amount spent: ₹{df['AmountSpent'].mean():.2f}")

# Q9. Count the number of users in each subscription/membership category
print("\nQ9. Customer count per Membership category:")
print(df['Membership'].value_counts())

# Q10. Determine the percentage of users who purchased (Proxy for renewal)
purchased_pct = df['Purchased'].value_counts(normalize=True) * 100
print("\nQ10. Percentage of users who completed a purchase:")
print(purchased_pct)

# Q11. Convert categorical features into numerical form
df_encoded = df.copy()
le_dict = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    le_dict[col] = le
print("Q11. Categorical columns encoded successfully.")

# Q12. Define the feature set (X) and target variable (y) for purchase prediction
# We drop 'CustomerID' (metadata), 'Purchased' (target), and 'AmountSpent' (leaks the answer)
X_clf = df_encoded.drop(['CustomerID', 'Purchased', 'AmountSpent'], axis=1)
y_clf = df_encoded['Purchased']
print("Q12. Features (X) and Target (y) defined for Classification.")

# Q13. Split the dataset into training and testing sets (80% train, 20% test)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)
print(f"Q13. Dataset split completed. Train shape: {X_train_c.shape}, Test shape: {X_test_c.shape}")

# Q14. Train a Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_c, y_train_c)

# Q15. Evaluate the model using accuracy
y_pred_dt = dt_model.predict(X_test_c)
print(f"Q15. Decision Tree Accuracy: {accuracy_score(y_test_c, y_pred_dt):.4f} ({accuracy_score(y_test_c, y_pred_dt)*100:.1f}%)")

# Q16. Generate and interpret the confusion matrix
print("\nQ16. Confusion Matrix:")
print(confusion_matrix(y_test_c, y_pred_dt))

# Q17. Train a KNN classifier with K = 5
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_c, y_train_c)

# Q18. Compare the accuracy of KNN with the Decision Tree model
y_pred_knn = knn_model.predict(X_test_c)
print(f"Q17 & Q18. KNN Accuracy (K=5): {accuracy_score(y_test_c, y_pred_knn):.4f} ({accuracy_score(y_test_c, y_pred_knn)*100:.1f}%)")

# Q19. Train a Linear Regression model to predict amount spent
X_reg = df_encoded.drop(['CustomerID', 'AmountSpent'], axis=1)
y_reg = df_encoded['AmountSpent']
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

lr_model = LinearRegression()
lr_model.fit(X_train_r, y_train_r)
print("Q19. Linear Regression model trained successfully.")

# Q20. Predict the monthly spending for a new user and interpret the result
new_user_sample = X_test_r.iloc[[0]]
predicted_spend = lr_model.predict(new_user_sample)
print(f"\nQ20. Predicted Amount Spent for the new sample customer: ₹{predicted_spend[0]:.2f}")
print(f"     Actual Amount Spent for this user was: ₹{y_test_r.iloc[0]:.2f}")
```
